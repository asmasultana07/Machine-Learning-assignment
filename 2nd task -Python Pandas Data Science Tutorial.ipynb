```python
bios['height_category'] = bios['height_cm'].apply(lambda x: 'Short'
                                                  if x < 165
                                                  else ('Average' if x < 185 else 'Tall'))
```

```python
def categorize_athlete(row):
    if row['height_cm'] < 175 and row['weight_kg'] < 70:
        return 'Lightweight'
    elif row['height_cm'] < 185 or row['weight_kg'] <= 80:
        return 'Middleweight'
    else:
        return 'Heavyweight'

bios['Category'] = bios.apply(categorize_athlete, axis=1)
```

## *Load & Accessing Dataset*

```python
import pandas as pd
import numpy as np

mobile1 = pd.read_csv('mobile_brands.csv')
mobile1.info()
```

```python
mobile1.head()
```

```python
mobile2 = pd.read_csv('mobile_sold.csv')
mobile2.info()
```

```python
mobile2.head()
```

```python
mobile2.sort_values(["year"], ascending=False)
```

```python
##rename columns to combine
aa= mobile2.rename(columns={'manufacturer':'Brand'})
aa.head()
```

## *Merge both dataset*

```python
new_mobile = pd.merge(mobile1,aa, on='Brand',how='left')
new_mobile.head()
```

```python
new_mobile.info()
```

```python
#sort the year
new_mobile= new_mobile.sort_values(["year"], ascending=False)
new_mobile.head()
```

## *Handling null values*

```python
bb= new_mobile[new_mobile['model'].isna()]
bb.info()
```

```python
#new_mobile.sample(5)
#bb.sample(5)
```

```python
df = new_mobile.dropna()
df.info()
```

```python
df.sample(6)
```

## *Aggregation data*

```python
df['Brand'].value_counts()
```

```python
df['Region'].value_counts()
```

```python
#df[df['Region']=='Asia']['Country'].value_counts().head(10)

df[df['Region']=='Asia']['Brand'].value_counts().head(10)
```

```python
#add a Price columns

```

```python
df.loc[:, 'price'] = np.where(df['Region'] == 'Asia', 11.11, 22.22)
df.sample(6)
```

```python
#groupby function

df.groupby(['Region', 'Brand']).agg({'units_sold_m': 'sum', 'price': 'mean'})
```

## *Advanced functionality*

```python
df['revenue'] = df['price'].shift(2)
df.head()
```

```python
#percentage
df['per_change'] = df['price']/ df['revenue'] * 100
df.head()
```

```python
df.sort_values(['units_sold_m'])
```

## ***Display***

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
avg_price_by_manufacturer = df.groupby('Brand')['price'].mean()

avg_price_by_manufacturer.plot(kind='bar', figsize=(10, 6), color='skyblue')
plt.title('Distribution of Average Price by Manufacturer Brand')
plt.xlabel('Brand')
plt.ylabel('Average Price')
#plt.grid(True)

plt.show()
```

```python
avg_price_by_manufacturer = df.groupby('Brand')['price'].mean()

# Plotting as a donut chart with percentages
avg_price_by_manufacturer.plot(
    kind='pie',
    figsize=(10, 6),
    colors=plt.cm.Paired.colors,  # This will give different colors to each slice
    wedgeprops={'width': 0.4},  # Adjust width to make it a donut chart
    autopct='%1.1f%%'  # Display percentages with 1 decimal place
)

plt.title('Distribution of Average Price by Manufacturer Brand')
plt.ylabel('')  # Remove the 'Average Price' label for clarity

plt.show()
```

```python

brand_units_sold = df.groupby('Brand')['units_sold_m'].mean().reset_index()

# Plotting the bar chart
plt.figure(figsize=(10,6))
plt.bar(brand_units_sold['Brand'], brand_units_sold['units_sold_m'], color='teal', edgecolor='black')

plt.title('Distribution of Units Sold by Mobile Brand')
plt.xlabel('Brand')
plt.ylabel('Units Sold')


plt.show()
```
