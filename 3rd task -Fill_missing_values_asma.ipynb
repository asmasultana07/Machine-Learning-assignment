## x groupby function

```python
bios1= bios_copy.copy()
bios1['height_cm']= bios1.height_cm.fillna(bios1.groupby(['born_country','NOC'])['height_cm'].transform('mean'))
bios1['weight_kg']= bios1.weight_kg.fillna(bios1.groupby(['born_country','NOC'])['weight_kg'].transform('mean'))
```

## x Interpolation (linear method)

group rows in a DataFrame by one or more columns

```python
#convert datetime
bios3['born_date'] = pd.to_datetime(bios3['born_date'], errors='coerce')
bios3['died_date'] = pd.to_datetime(bios3['died_date'], errors='coerce')

bios3['born_date'] = bios3['born_date'].interpolate(method='linear')
bios3['died_date'] = bios3['died_date'].interpolate(method='linear')

bios3.head(14)
```

# Q3. Use ML model to predict the missing values

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
from sklearn.ensemble import RandomForestRegressor

biosmodel= bios_copy.copy()
biosmodel = biosmodel[['athlete_id','height_cm','weight_kg']]
```

```python
imp= IterativeImputer(RandomForestRegressor(),max_iter=10,random_state=0)
model= pd.DataFrame(imp.fit_transform(biosmodel),columns = biosmodel.columns)
model.head(14)
```
