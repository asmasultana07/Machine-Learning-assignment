# Second Tutorial

## import data

```python
import pandas as pd
cc= pd.read_csv('https://raw.githubusercontent.com/dataprofessor/data/master/delaney_solubility_with_descriptors.csv')
cc
```

## data preparation

```python
y=cc['logS']
```

```python
x=cc.drop('logS',axis=1)
```

```python
#data splitting
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=100)
x_train
```

```python
x_test
```

# *Linear* *Regression*

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train, y_train)
```

```python
y_lr_train_pred=lr.predict(x_train)
y_lr_test_pred=lr.predict(x_test)
```

```python
y_train
print(y_lr_train_pred)
```

### evaluate model perfomance

```python
from sklearn.metrics import mean_squared_error,r2_score
lr_train_maen=mean_squared_error(y_train,y_lr_train_pred)
lr_train_r2=r2_score(y_train,y_lr_train_pred)

lr_test_maen=mean_squared_error(y_test,y_lr_test_pred)
lr_test_r2=r2_score(y_test,y_lr_test_pred)
```

```python
lr_train_maen
```

```python
print("lr mse (train): ",lr_train_maen)
print("lr r2 (train): ",lr_train_r2)
print("lr mse (test): ",lr_test_maen)
print("lr r2 (test): ",lr_test_r2)
```

```python
lr_resuls= pd.DataFrame(['Linear Regression',lr_train_maen,lr_train_r2,lr_test_maen,lr_test_r2]).transpose()
lr_resuls.columns=['Method','Training MSE','Training R2','Test MSE','Test R2']
lr_resuls
```

## LR visualization

```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(5,5))
plt.scatter(x=y_train, y=y_lr_train_pred, c='#40E0D0', alpha=0.3)

z= np.polyfit(y_train, y_lr_train_pred, 1)
p = np.poly1d(z)

plt.plot(y_train, p(y_train), '#F8766D')
plt.ylabel('Predict LogS')
plt.xlabel('Experiment LogS')
```

# *Random Forest*

```python
from sklearn.ensemble import RandomForestRegressor
rf= RandomForestRegressor(max_depth=2,random_state=100)
rf.fit(x_train,y_train)
```

```python
y_rf_train_pred= rf.predict(x_train)
y_rf_test_pred= rf.predict(x_test)
```

### evaluate model perfomance

```python
from sklearn.metrics import mean_squared_error,r2_score

rf_train_maen= mean_squared_error(y_train,y_rf_train_pred)
rf_train_r2= r2_score(y_train,y_rf_train_pred)

rf_test_maen= mean_squared_error(y_test,y_rf_test_pred)
rf_test_r2= r2_score(y_test,y_rf_test_pred)
```

```python
rf_resuls= pd.DataFrame(['Random Forest',rf_train_maen,rf_train_r2,rf_test_maen,rf_test_r2]).transpose()
rf_resuls.columns=['Method','Training MSE','Training R2','Test MSE','Test R2']
rf_resuls
```

## Model comparison ***LR vs RF***

```python
models= pd.concat([lr_resuls,rf_resuls], axis=0)
models
```

## RF visualization

```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(5,5))
plt.scatter(x=y_train, y=y_rf_train_pred, c='#40E0D0', alpha=0.3)

z= np.polyfit(y_train, y_rf_train_pred, 1)
p = np.poly1d(z)

plt.plot(y_train, p(y_train), '#F8766D')
plt.ylabel('Predict LogS')
plt.xlabel('Experiment LogS')
```

#  KNN- k nearest neighbors

```python
from sklearn.neighbors import KNeighborsRegressor
knn = KNeighborsRegressor(n_neighbors=10)
knn.fit(x_train, y_train)
```

```python
y_knn_train_pred=knn.predict(x_train)
y_knn_test_pred=knn.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error,r2_score

knn_train_maen= mean_squared_error(y_train,y_knn_train_pred)
knn_train_r2= r2_score(y_train,y_knn_train_pred)

knn_test_maen= mean_squared_error(y_test,y_knn_test_pred)
knn_test_r2= r2_score(y_test,y_knn_test_pred)
```

```python
knn_results= pd.DataFrame(['KNN',knn_train_maen,knn_train_r2,knn_test_maen,knn_test_r2]).transpose()
knn_results.columns=['Method','Training MSE','Training R2','Test MSE','Test R2']
knn_results
```

## LR vs RF vs KNN

```python
modelss= pd.concat([lr_resuls,rf_resuls,knn_results], axis=0)
modelss
```
